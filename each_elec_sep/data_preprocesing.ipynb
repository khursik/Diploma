```python
import pandas as pd
```

```python
df = pd.read_csv("~\PycharmProjects\Diploma\EDA\data.csv")
df.head()
```

```python
df.shape
```

```python
# Let's average the frequencies over four ranges

avg_vals = df.iloc[:, :-2:4].values

for i in range(1, 4):
    avg_vals += df.iloc[:, i:-2:4].values
avg_vals /= 4

avg_vals.shape
```

```python
new_df = pd.DataFrame(avg_vals, columns=[x for x in df.columns[:-2] if int(x.split('_')[1]) < 7])
new_df['name'] = df['name']
new_df['task'] = df['task']
```

```python
new_df
```

```python
new_df.to_csv("data\with_avg_vals.csv", index=False)
```
